## Conversão dos arquivos de CSV para parquet

```python
# Converter o CSV calendar para parquet

df = pd.read_csv("calendar.csv")

df.to_parquet("calendar.parquet", index=False)
```

```python
# Converter o CSV listings para parquet

df = pd.read_csv("listings.csv")

df.to_parquet("listings.parquet", index=False)
```

```python
# Converter o CSV reviews para parquet

df = pd.read_csv("reviews.csv")

df.to_parquet("reviews.parquet", index=False)
```
