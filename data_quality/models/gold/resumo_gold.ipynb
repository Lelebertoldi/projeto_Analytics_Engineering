## Tabela demanda_por_bairro:

### Colunas e tipo de colunas

```python
# Conectar ao banco de dados
conn = duckdb.connect(database=db_path)

# Consulta SQL para pegar todas as colunas e seus tipos da tabela 'demanda_por_bairro'
query = """
SELECT column_name, data_type
FROM information_schema.columns
WHERE table_name = 'demanda_por_bairro'
ORDER BY ordinal_position;
"""

# Executar a consulta
result = conn.execute(query).fetchall()

# Fechar a conexão
conn.close()

# Exibir os resultados
for row in result:
    print(f"Coluna: {row[0]}, Tipo: {row[1]}")
```

### Dataframe dos bairros com maior e menor demanda por mês:

```python
# Conectar ao banco DuckDB
con_gold = duckdb.connect(database=db_path)

# Consultar os dados da tabela demanda_por_bairro
query = """
    SELECT *
    FROM demanda_por_bairro
"""

# Executar a consulta e salvar os resultados em um DataFrame
df_demanda = con_gold.execute(query).fetchdf()

# Fechar a conexão com o banco de dados
con_gold.close()

# Listar os meses
meses = ['jan', 'fev', 'mar', 'abr', 'mai', 'jun', 'jul', 'ago', 'set', 'out', 'nov', 'dez']

# Criar uma lista para armazenar os resultados
resultados = []

# Para cada mês, encontrar o bairro com maior e menor demanda
for mes in meses:
    coluna_mes = f'demanda_{mes}'
    maior_demanda_bairro = df_demanda.loc[df_demanda[coluna_mes].idxmax()]['bairro']
    menor_demanda_bairro = df_demanda.loc[df_demanda[coluna_mes].idxmin()]['bairro']
    
    # Adicionar os resultados na lista
    resultados.append({
        'Mês': mes.capitalize(),
        'Bairro com Maior Demanda': maior_demanda_bairro,
        'Bairro com Menor Demanda': menor_demanda_bairro
    })

# Criar o DataFrame com os resultados
df_resultados = pd.DataFrame(resultados)

# Exibir o DataFrame
display(df_resultados)
```

### Bairro com maior preço, menor preço, maior avaliação, menor avaliação:

```python
# Conectar ao banco DuckDB
con_gold = duckdb.connect(database=db_path)

# Consultar os dados da tabela demanda_por_bairro
query = """
    SELECT bairro, media_preco, media_avaliacoes
    FROM demanda_por_bairro
"""

# Executar a consulta e salvar os resultados em um DataFrame
df_demanda = con_gold.execute(query).fetchdf()

# Fechar a conexão com o banco de dados
con_gold.close()

# Encontrar o bairro com maior preço
bairro_maior_preco = df_demanda.loc[df_demanda['media_preco'].idxmax()]['bairro']
maior_preco = df_demanda['media_preco'].max()

# Encontrar o bairro com menor preço
bairro_menor_preco = df_demanda.loc[df_demanda['media_preco'].idxmin()]['bairro']
menor_preco = df_demanda['media_preco'].min()

# Encontrar o bairro com maior média de avaliações
bairro_maior_avaliacao = df_demanda.loc[df_demanda['media_avaliacoes'].idxmax()]['bairro']
maior_avaliacao = df_demanda['media_avaliacoes'].max()

# Encontrar o bairro com menor média de avaliações
bairro_menor_avaliacao = df_demanda.loc[df_demanda['media_avaliacoes'].idxmin()]['bairro']
menor_avaliacao = df_demanda['media_avaliacoes'].min()

# Criar o DataFrame com os resultados
df_resultados = pd.DataFrame({
    'Métrica': ['Maior Preço', 'Menor Preço', 'Maior Avaliação', 'Menor Avaliação'],
    'Bairro': [bairro_maior_preco, bairro_menor_preco, bairro_maior_avaliacao, bairro_menor_avaliacao],
    'Valor': [maior_preco, menor_preco, maior_avaliacao, menor_avaliacao]
})

# Exibir o DataFrame
display(df_resultados)
```
